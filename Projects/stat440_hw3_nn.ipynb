```python
import tensorflow as tf
from tensorflow import keras
fashion_mnist = keras.datasets.fashion_mnist
(tri, trl), (tei, tel) = fashion_mnist.load_data()
tri = tri / 255.0
tei = tei / 255.0
model = keras.Sequential([
  keras.layers.Flatten(input_shape=(28, 28)),
  keras.layers.Dense(128, activation=tf.nn.relu),
  keras.layers.Dense(10, activation=tf.nn.softmax)])
model.compile(optimizer=tf.optimizers.Adam(),
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])
model.fit(tri, trl, epochs = 5)
tl, ta = model.evaluate(tei, tel)
print('Test accuracy: %0.3f (%% correct)' % (ta * 100))
```

```python
from tensorflow import keras
from keras.utils import to_categorical
fashion_mnist = keras.datasets.fashion_mnist
(tri, trl), (tei, tel) = fashion_mnist.load_data()

tri = tri.reshape((60000, 28, 28, 1))
tei = tei.reshape((10000, 28, 28, 1))

tri = tri / 255.0
tei = tei / 255.0

trl = to_categorical(trl)
tel = to_categorical(tel)

model = keras.Sequential([
    keras.layers.Conv2D(128, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')])

model.compile(optimizer=tf.optimizers.Adam(),
  loss='categorical_crossentropy',
  metrics=['accuracy'])

model.fit(tri, trl, epochs = 5, batch_size = 128)

tl, ta = model.evaluate(tei, tel)
print('Test accuracy: %0.3f (%% correct)' % (ta * 100))
```
